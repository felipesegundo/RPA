```python
# PACOTES NECESSERAIOS
import pandas as pd
import requests
```

```python
# URL DA PAGINA ONDE TEM A TABELA
url = 'https://br.investing.com/indices/bovespa-components'
```

```python
request = requests.get(url)
```

```python
df = pd.read_html(request.text)
```

```python
df
```

```python
# ORDENAÇÃO DAS TABELAS POR INDICE
for i, tabela in enumerate(df):
    print("----------------------------------")
    print(i)
    print(tabela)
```

```python
# ESCOLHA DO INDICE ONDE CONTEM A TABELA DESEJADA
df = pd.DataFrame(df[0])
```
